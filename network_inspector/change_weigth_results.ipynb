```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
err_dict = {}
for layer in 0, 3, 6, 8, 10:
    errors = []
    with open(f"features{layer}.txt") as f:
        for line in f:
            if "Got" in line:
                error = float(line.split(" ")[-1])
                errors.append(error)
    err_dict[layer] =  errors[2:]            
```

```python
df = pd.DataFrame(err_dict)

fig, ax = plt.subplots(figsize=(15,3))
sns.violinplot(data=df, ax=ax);

fig, axs = plt.subplots(figsize=(15,3), ncols=5)
sns.violinplot(data=df[0], ax=axs[0])
sns.violinplot(data=df[3], ax=axs[1])
sns.violinplot(data=df[6], ax=axs[2])
sns.violinplot(data=df[8], ax=axs[3])
sns.violinplot(data=df[10], ax=axs[4])

```
