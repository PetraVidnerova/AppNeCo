```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("alexnet_cifar_outputs.parquet")
df
```

```python
min_val = int(df["value"].min())-1
max_val = int(df["value"].max())+1
min_val, max_val
```

```python
g = sns.FacetGrid(df, row='class', col='output', height=5, aspect=2)
g.map(sns.histplot, 'value', bins=range(min_val, max_val, 1), kde=False)
plt.xlabel('Output value')
plt.ylabel('Frequency')
plt.legend()
plt.xticks(range(min_val, max_val, 20))
plt.show()
```
